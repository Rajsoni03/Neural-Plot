# 2 - Neural Net Visualization 

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

## Sample Layers

```python
layers1 = [(12, 12, 3),
          (8, 8, 2),
          (8,  8, 3),
          (4, 4),
          (2,)]

layers2 = [(12, 12, 3),
          (8, 8, 2),
          (8,  8, 3),]
```

## Testing 1

```python
%matplotlib notebook
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

layers = [(12, 12, 3),(8, 8, 2),(5, 5, 3),(4, 4, 3)]
len_z = 0

for i in layers:
    try:
        len_z += i[2]
    except:
        len_z += 1
        
sx, sy, sz = [], [], []

last_temp = 0

for lay in layers2:
    x = lay[0]
    y = lay[1]
    z = lay[2]
    
    for k in range(z):
        for i in range(y):
            sx += [j for j in range(x)]

        for i in range(y):
            sy += [j for j in (np.ones(x,dtype=int) * i)]

        for i in range(x):
            sz += [j for j in (np.ones(y,dtype=int) * last_temp)]

        last_temp += 1    

ax.scatter(sx, sz, sy, c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Z Label')
ax.set_zlabel('Y Label')

plt.show()
```

## Testing 2

```python
%matplotlib notebook
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

layers = [(12, 12, 3),(8, 8, 2),(5, 5, 3),(4, 4, 3)]
len_z = 0

for i in layers:
    try:
        len_z += i[2]
    except:
        len_z += 1
        
sx, sy, sz = [], [], []

last_temp = 0

all_xyz = []
x_max, y_max = 0,0

for lay in layers2:
    x = lay[0]
    y = lay[1]
    z = lay[2]
    
    if x_max < x:
        x_max = x
    if y_max < y:
        y_max = y
    
    single_layer = []
    for k in range(z):
        mx,my,mz = [],[],[]
        for i in range(y):            
            ox = [j for j in range(x)]
            sx += ox
            mx.append(ox)

        for i in range(y):
            oy = [j for j in (np.ones(x,dtype=int) * i)]
            sy += oy
            my.append(oy)

        for i in range(x):
            oz = [j for j in (np.ones(y,dtype=int) * last_temp)]
            sz += oz
            mz.append(oz)

        last_temp += 1  
        
        single_layer.append([mx,my,mz])
    all_xyz.append(single_layer)

ax.scatter(sx, sz, sy, c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Z Label')
ax.set_zlabel('Y Label')

plt.show()
```

## Creating A Ploting function

```python
def neuralplot(layers):
    
    %matplotlib notebook
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection='3d')

    sx, sy, sz = [], [], []

    x_max, y_max = 0,0
    last_layer = 0
    all_xyz = []
    all_c, all_marker = [], []
    for lay in layers:
        x = lay[0]
        try:
            y = lay[1]
        except:
            y = 1
        try:
            z = lay[2]
        except:
            z = 1
        try:
            all_c.append(lay[3])
        except:
            all_c.append('r')
        try:
            all_marker.append(lay[4])
        except:
            all_marker.append('o')

        if x_max < x:
            x_max = x
        if y_max < y:
            y_max = y

        single_layer = []
        for k in range(z):
            mx,my,mz = [],[],[]
            for i in range(y):            
                ox = [j for j in range(x)]
                mx.append(ox)

            for i in range(y):
                oy = [j for j in (np.ones(x,dtype=int) * i)]
                my.append(oy)

            for i in range(x):
                oz = [j for j in (np.ones(y,dtype=int) * last_layer)]
                mz.append(oz)
            last_layer += 1  
            single_layer.append([mx,my,mz])
        all_xyz.append(single_layer)
        last_layer += 4  
    

    for i, c, marker in zip(all_xyz, all_c, all_marker):
        x, y, z = [],[],[]
        if c == 'rgb' and len(i) == 3:
            for j, color in zip(i, ['r','g','b']):
                x, y, z = [],[],[]                
                for k in j[0]:
                    k = [a + ((x_max-len(k))/2) for a in k]                
                    x += k
                for l in j[1]:
                    l = [b + ((y_max-(j[1][-1][-1]+1))/2) for b in l]
                    y += l
                for k in j[2]:
                    z += k
                ax.scatter(x, z, y, c=color, marker=marker, s=5)
                    
        else:
            for j in i:
                x, y, z = [],[],[]                
                for k in j[0]:
                    k = [a + ((x_max-len(k))/2) for a in k]                
                    x += k
                for l in j[1]:
                    l = [b + ((y_max-(j[1][-1][-1]+1))/2) for b in l]
                    y += l
                for k in j[2]:
                    z += k
                ax.scatter(x, z, y, c=c, marker=marker, s=5)    
    ax.set_xlabel('X-Axis')
    ax.set_ylabel('Z-Axis')
    ax.set_zlabel('Y-Axis')

    plt.show()
```

##  Call neuralplot() Function and plot 3D graph

```python
layers = [(12, 12, 3, 'rgb'),(8, 8, 2, 'b', '^'),(5, 5, 3, 'g'),(4, 4, 3, 'y', '^')]
neuralplot(layers)
```
