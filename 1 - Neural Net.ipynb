## Import Libaries

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Define a function for creating 2D Matrix 

```python
def mat2d(x,y=1):
    sx, sy = [], []
    for i in range(y):
        sx += [j for j in range(x)]
    for i in range(y):
        sy += [j for j in (np.ones(x,dtype=int) * i)]
    return sx, sy


x,y = mat2d(4,3)
print(x)
print(y)
print(len(x),len(y))


    
```

## 2D Scatter Plot 

```python
x,y = mat2d(4,5)
plt.scatter(x,y)
plt.show()
```

## Define function for creating 3D Matrix 

```python
def mat3d(x,y=1,z=1):
    sx, sy, sz = [], [], []
    
    for k in range(z):
        for i in range(y):
            sx += [j for j in range(x)]

        for i in range(y):
            sy += [j for j in (np.ones(x,dtype=int) * i)]

        for i in range(x):
            sz += [j for j in (np.ones(y,dtype=int) * k)]
        
    return sx, sy, sz


x,y,z = mat3d(4,3,2)
print(x)
print(y)
print(z)
print(len(x),len(y),len(z))


    
```

## 3D Scatter Plot 

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

x,y,z = mat3d(32, 32, 3)

ax.scatter(x, z, y, c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Z Label')
ax.set_zlabel('Y Label')

plt.show()
```

```python
def mat3dl(x,y=1,z=1,layer=None):
    sx, sy, sz = [], [], []
    
    for k in range(z):
        for i in range(y):
            sx += [j for j in range(x)]

        for i in range(y):
            sy += [j for j in (np.ones(x,dtype=int) * i)]

        for i in range(x):
            sz += [j for j in (np.ones(y,dtype=int) * k)]
    
    if layer is not None:
        slice_from  = layer[0] * (x*y)
        slice_to =  layer[1] * (x*y)
        sl = slice(slice_from, slice_to, None)
        return sx[sl], sy[sl], sz[sl]
    return sx, sy, sz


x,y,z = mat3dl(2,2,3, (2,3))
print(x)
print(y)
print(z)
print(len(x),len(y),len(z))


    
```

## Advanced 3D Scatter Plot for RGB layer Visualization

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

for i, c in enumerate(['r', 'g', 'b']):
    x,y,z = mat3dl(32, 32, 3, (i,i+1))
    ax.scatter(x, z, y, c=c, marker='o')
    
ax.set_xlabel('X Label')
ax.set_ylabel('Z Label')
ax.set_zlabel('Y Label')

plt.show()
```
